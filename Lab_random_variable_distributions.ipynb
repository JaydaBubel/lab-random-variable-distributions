1.Get the numerical variables from our dataset.

2.Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

3.Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Ironhack/we_fn_use_c_marketing_customer_value_analysis-6.csv')
```

```python
numerical = df.select_dtypes(include=[np.number])
numerical
```

```python
for column in numerical.columns:
    plt.figure(figsize=(8, 4))

    sns.histplot(numerical[column], kde=True, bins=30)

    plt.title(f'distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('frequency')
    plt.show()
```

```python
#Check if transformations (log-transform, etc.) up to this point changes result.
for column in numerical.columns:
    plt.figure(figsize=(8, 4))

    log_transformed = np.log1p(numerical[column])

    sns.histplot(log_transformed, kde=True, bins=30)

    plt.title(f'Log-Transformed Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
